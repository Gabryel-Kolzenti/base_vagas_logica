```python
def base_painel_prod(df_base_painel_prod):
    
    

    # Filtragem dos DataFrames
    filter_strtype = [30]
    filter_strtype2 = [31]
    filter_strtype3 = [42]
    filter_strtype4 = [50]
    df_prod_filtered   = df_base_painel_prod.loc[df_base_painel_prod['ID_INDICADOR'].isin([30])].copy()
    df_margin_filtered = df_base_painel_prod.loc[df_base_painel_prod['ID_INDICADOR'].isin([31])].copy()
    df_desc_filtered   = df_base_painel_prod.loc[df_base_painel_prod['ID_INDICADOR'].isin([42])].copy()
    df_lb_filtered     = df_base_painel_prod.loc[df_base_painel_prod['ID_INDICADOR'].isin([50])].copy()


    # Criação da coluna "DESCONTO" no DataFrame df_margin_filtered
    df_desc_filtered.loc[:,'DESC'] = df_desc_filtered.loc[:,'REAL_MES']
    df_desc_filtered.loc[:,'DESC']=pd.to_numeric(df_desc_filtered.loc[:,'DESC'].replace(',', '.'), errors='coerce')

    # Criação da coluna "LUCRO BRUTO" no DataFrame df_margin_filtered
    df_lb_filtered.loc[:,'LB_REAL'] = df_lb_filtered.loc[:,'REAL_MES']
    df_lb_filtered.loc[:,'LB_ORC'] = df_lb_filtered.loc[:,'ORC_MES']
    df_lb_filtered.loc[:,'LB_REAL']=pd.to_numeric(df_lb_filtered.loc[:,'LB_REAL'].replace(',', '.'), errors='coerce')
    df_lb_filtered.loc[:,'LB_ORC']=pd.to_numeric(df_lb_filtered.loc[:,'LB_ORC'].replace(',', '.'), errors='coerce')
    df_lb_filtered.loc[:,'LB_ATING'] = df_lb_filtered.apply(lambda row: row['LB_REAL'] / row['LB_ORC'] if row['LB_ORC'] != 0 else 0, axis=1)       

    # Criação da coluna "MARGEM" no DataFrame df_margin_filtered
    df_margin_filtered.loc[:,'MARGEM'] = df_margin_filtered.loc[:,'REAL_MES']
    df_margin_filtered.loc[:,'MARGEM']=pd.to_numeric(df_margin_filtered.loc[:,'MARGEM'].replace(',', '.'), errors='coerce')
 
    # Conversão das colunas para números
    numeric_columns = ['REAL_MES', 'ORC_MES', 'META_MES']
    df_prod_filtered[numeric_columns] = df_prod_filtered[numeric_columns].apply(lambda x: pd.to_numeric(x.replace(',', '.'), errors='coerce'))
 
    # Cálculo das colunas "ATING" e "PROJ" no DataFrame df_prod_filtered
    df_prod_filtered.loc[:,'ATING'] = df_prod_filtered.apply(lambda row: row['REAL_MES'] / row['ORC_MES'] if row['ORC_MES'] != 0 else 0, axis=1)
    df_prod_filtered.loc[:,'PROJ'] = df_prod_filtered.loc[:,'ATING'] * df_prod_filtered.loc[:,'META_MES']

    # Mesclagem dos DataFrames
    df_prod_final = pd.merge(df_prod_filtered, df_margin_filtered[['FILIAL', 'MARGEM']], on='FILIAL', how='left')
    df_prod_final = pd.merge(df_prod_final,df_desc_filtered[['FILIAL','DESC']],on='FILIAL',how='left')
    df_prod_final = pd.merge(df_prod_final,df_lb_filtered[['FILIAL','LB_REAL','LB_ATING']],on='FILIAL',how='left')

    df_prod_final.loc[:,'REAL_DIA']=pd.to_numeric(df_prod_final.loc[:,'REAL_DIA'].replace(',', '.'), errors='coerce')   
    df_prod_final = df_prod_final[['FILIAL', 'REAL_DIA', 'REAL_MES', 'ATING', 'PROJ', 'META_MES','MARGEM','DESC', 'LB_REAL','LB_ATING','ORC_MES']]

    return df_prod_final
```
