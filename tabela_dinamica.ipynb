## Tentativa de utilização de dados de tabela dinâmica

```python
import pandas as pd
import xlwings as xw
```

```python

arquivo = r"O:\CONTROLADORIA\Ticket Médio\Relatórios Produção Controladoria 05-2025.xlsb"

df = pd.read_excel(arquivo, header=None)
df.columns = df.iloc[10]

print(df.columns)
df = df.drop(index=range(0, 11)).reset_index(drop=True)
df.columns = df.columns.str.replace('\n', ' ').str.strip()

# Selecionar colunas de interesse
col_filial = 'Regionais/Filiais'
col_valor = 'Total Ano Anterior Prod'

# Criar colunas auxiliares
df['Região'] = None

regiao_atual = None
for i, valor in df[col_filial].items():
    if isinstance(valor, str) and valor.strip() != "":
        if valor.startswith('0') or valor[0].isdigit():  # Linha de uma nova Região
            regiao_atual = valor.strip()
        else:
            df.at[i, 'Região'] = regiao_atual  # Atribui a região à filial

# Filtrar apenas as linhas que são filiais (não a linha da região em si, mas sim os subgrupos)
df_filiais = df[df['Região'].notna()][[col_filial, 'Região', col_valor]]

# Exibir resultado
print(df_filiais)
```
